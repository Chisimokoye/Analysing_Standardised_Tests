```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_path = "../PyCitySchool/schools_complete.csv"
student_data_path = "../PyCitySchool/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_data_path)
student_df = pd.read_csv(student_data_path)

school_df.head()
```

```python
student_df.head()
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## DISTRICT SUMMARY

```python
#Calculate the total number of schools
Total_num_of_schools = school_data_complete['school_name'].nunique()

#Calculate the total number of students
Total_num_of_students = school_data_complete['student_name'].count()

#Calculate the total budget
budget = school_data_complete['budget'].unique()
Total_budget = budget.sum()
Total_budget

#Calculate the average maths score
Average_maths_score = school_data_complete['math_score'].mean()

#Calculate the average reading score
Average_reading_score = school_data_complete['reading_score'].mean()

#Calculate the percentage of students with a passing math score (70 or greater)
passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
number_passing_math = passing_math['math_score'].count()
percentage_passing_math = number_passing_math/Total_num_of_students * 100

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
number_passing_reading = passing_reading['reading_score'].count()
percentage_passing_reading = number_passing_reading/Total_num_of_students * 100

#Calculate the percentage of students who passed math and reading (% Overall Passing)
percentage_overall_passing = ((percentage_passing_math/100) * (percentage_passing_reading/100)) *100
```

```python
#Create a dataframe to hold the above results
district_summary = pd.DataFrame({'Total Schools': [Total_num_of_schools],
                                'Total students': [Total_num_of_students],
                                'Total budget': [Total_budget],
                                'Average Math Score': [Average_maths_score],
                                'Average Reading Score': [Average_reading_score],
                                '% Passing Math': [percentage_passing_math],
                                '% Passing Reading': [percentage_passing_reading],
                                '% Overall Passing': [percentage_overall_passing]})

#Optional: give the displayed data cleaner formatting
district_summary['Total students'] = district_summary['Total students'].map("{:,.0f}".format)
district_summary['Total budget'] = district_summary['Total budget'].map("${:,.2f}".format)
district_summary['Average Math Score'] = district_summary['Average Math Score'].map("{:.2f}".format)
district_summary['Average Reading Score'] = district_summary['Average Reading Score'].map("{:.2f}".format)
district_summary['% Passing Math'] = district_summary['% Passing Math'].map("{:.2f}".format)
district_summary['% Passing Reading'] = district_summary['% Passing Reading'].map("{:.2f}".format)
district_summary['% Overall Passing'] = district_summary['% Overall Passing'].map("{:.2f}".format)

district_summary
```

## School Summary

```python
#SCHOOL SUMMARY
school_summary = school_data_complete.groupby(['school_name'])
school_summary_df = school_summary.sum()
school_summary_df.head()
```

```python
#Create an overview table that summarizes key metrics about each school
#School_type
school_type = school_data_complete.groupby('school_name')['type'].first()

#Total_students 
total_students_sum = school_data_complete.groupby('school_name')['student_name'].count()

#Total school budget
total_school_budget = school_data_complete.groupby('school_name')['budget'].first()

#Per Student Budget
per_student_budget = total_school_budget/total_students_sum

#Average Math Score
Average_maths_score = school_data_complete.groupby('school_name')['math_score'].mean()

#Average Reading Score
Average_reading_score = school_data_complete.groupby('school_name')['reading_score'].mean()

#% Passing Math
passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70].groupby('school_name')
number_passing_math = passing_math['math_score'].count()
percentage_passing_math = number_passing_math/total_students_sum * 100

#% Passing Reading
passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70].groupby('school_name')
number_passing_reading = passing_reading['reading_score'].count()
percentage_passing_reading = number_passing_reading/total_students_sum * 100

#% Overall Passing (The percentage of students that passed math and reading.)
percentage_overall_passing = ((percentage_passing_math/100) * (percentage_passing_reading/100)) *100
```

```python
#Create a dataframe to hold the above results
school_summary_df['School Type']=school_type
school_summary_df['Total Students']=total_students_sum
school_summary_df['Total School Budget']= total_school_budget
school_summary_df['Per Student Budget']= per_student_budget
school_summary_df['Average Math Score']= Average_maths_score
school_summary_df['Average Reading Score']= Average_reading_score
school_summary_df['% Passing Math']= percentage_passing_math
school_summary_df['% Passing Reading']= percentage_passing_reading
school_summary_df['% Overall Passing']= percentage_overall_passing

school_summary_new = school_summary_df.drop(columns=['Student ID', 'reading_score', 'math_score', 'School ID',
                                'size', 'budget'])

#Optional: give the displayed data cleaner formatting
school_summary_new['Total Students'] = school_summary_new['Total Students'].map("{:,.0f}".format)
school_summary_new['Total School Budget'] = school_summary_new['Total School Budget'].map("${:,.2f}".format)
school_summary_new['Per Student Budget'] = school_summary_new['Per Student Budget'].map("${:,.2f}".format)
school_summary_new['Average Math Score'] = school_summary_new['Average Math Score'].map("{:.3f}".format)
school_summary_new['Average Reading Score'] = school_summary_new['Average Reading Score'].map("{:.3f}".format)
school_summary_new['% Passing Math'] = school_summary_new['% Passing Math'].map("{:.3f}".format)
school_summary_new['% Passing Reading'] = school_summary_new['% Passing Reading'].map("{:.3f}".format)
school_summary_new['% Overall Passing'] = school_summary_new['% Overall Passing'].map("{:.3f}".format)

school_summary_new
```

# Scores by School Spending

```python
#Create a table that breaks down school performances based on average Spending Ranges (Per Student). 
school_summary_df.head()
```

```python
# Use 4 reasonable bins to group school spending
bins = [0, 585, 630, 645, 680]
scores_labels = ['< $585', '$585-630', '$630-645', '$645-680']

pd.cut(school_summary_df['Per Student Budget'], bins, labels = scores_labels)
school_summary_df['Spending Ranges (Per Student)'] = pd.cut(school_summary_df ['Per Student Budget'], bins, labels = scores_labels)
school_summary_df.head()
```

```python
scores_by_spending = school_summary_df.groupby('Spending Ranges (Per Student)')
scores_by_spending
```

```python
#Average Math Score
#Average Reading Score
#% Passing Math
#% Passing Reading
#Overall Passing Rate (Average of the above two)
scores_final = scores_by_spending[['Average Math Score', 'Average Reading Score',
                '% Passing Math', '% Passing Reading','% Overall Passing']].mean()

scores_final['Average Math Score'] = scores_final['Average Math Score'].map("{:.3f}".format)
scores_final['Average Reading Score'] = scores_final['Average Reading Score'].map("{:.3f}".format)
scores_final['% Passing Math'] = scores_final['% Passing Math'].map("{:.3f}".format)
scores_final['% Passing Reading'] = scores_final['% Passing Reading'].map("{:.3f}".format)
scores_final['% Overall Passing'] = scores_final['% Overall Passing'].map("{:.3f}".format)
scores_final
```
